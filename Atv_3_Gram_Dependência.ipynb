```python
pattern_chega = [
    # anchor token: chega
    { 
      "RIGHT_ID": "root",
      "RIGHT_ATTRS": { "ORTH": "chega" }
    },
    {
      "LEFT_ID": "root",
      "REL_OP": ">",
      "RIGHT_ID": "nsubj_conf",
      "RIGHT_ATTRS": { "DEP" : "nsubj", "ORTH": "confissões" }
    },
    {
      "LEFT_ID": "nsubj_conf",
      "REL_OP": ">",
      "RIGHT_ID": "punct_<<",
      "RIGHT_ATTRS": { "DEP" : "punct", "ORTH": "<<" }
    },
    {
      "LEFT_ID": "nsubj_conf",
      "REL_OP": ">",
      "RIGHT_ID": "punct_>>",
      "RIGHT_ATTRS": { "DEP" : "punct", "ORTH": ">>" }
    },
    {
      "LEFT_ID": "root",
      "REL_OP": ">",
      "RIGHT_ID": "obl_pt",
      "RIGHT_ATTRS": { "DEP" : "obl", "ORTH": "Portugal" }
    },
    {
      "LEFT_ID": "obl_pt",
      "REL_OP": ">",
      "RIGHT_ID": "case_obl",
      "RIGHT_ATTRS": { "DEP" : "case", "ORTH": "a" }
    },
]
```

```python
from spacy.matcher import DependencyMatcher
```

```python
matcher = DependencyMatcher(nlp.vocab)
matcher.add("estreia", [pattern_estreia])
doc = nlp("Manchete estréia novo jornalístico")
matches = matcher(doc)
print(matches)
match_id, token_ids = matches[0]
for i in range(len(token_ids)):
    print(pattern_estreia[i]["RIGHT_ID"] + ":", doc[token_ids[i]].text)
```

```python
matcher = DependencyMatcher(nlp.vocab)
matcher.add("joga", [pattern_joga])
doc = nlp("Aqui só joga quem está bem")
matches = matcher(doc)
print(matches)
```

```python
matcher = DependencyMatcher(nlp.vocab)
matcher.add("chega", [pattern_chega])
doc = nlp("«Confissões» chega a Portugal")
matches = matcher(doc)
print(matches)
```

```python
match_id, token_ids = matches[0]
for i in range(len(token_ids)):
    print(pattern_joga[i]["RIGHT_ID"] + ":", doc[token_ids[i]].text)
```

##Conllu

```python
from io import open
from conllu import parse_tree_incr

sentences = []

data_file = open(filename, "r", encoding="utf-8")
for tokentree in parse_tree_incr(data_file):
    sentences.append(tokentree)

#print(sentences)
```

```python
for sentence in sentences:
    root = sentence
    root.print_tree()
    print('\n')
```

##Displacy

```python
options = {"font": "Source Sans Pro",
           "fine_grained": True}
```

```python
for sentence in sentences:
    text_to_print_tree = sentence.metadata['text']
    print(text_to_print_tree)
    prep_text = nlp(text_to_print_tree)
    displacy.render(prep_text, style='dep', jupyter=True, options=options)
```
